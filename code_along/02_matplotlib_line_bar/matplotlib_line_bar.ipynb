```python
import duckdb

df_quantity_brand = duckdb.query("""
    Select make as Brand, SUM(quantity) as Quantity
    From df
    GROUP BY make
    Order By Quantity DESC
""").df()
```

## Bar chart 

```python
df_quantity_brand.head().plot(kind = "bar", x = "Brand", y= "Quantity", title="New Car Sales", xlabel="Brand", ylabel="Quantity")
```

## Matplotlib bar

```python
import matplotlib.pyplot as plt

df_quantity_brand_top10 = df_quantity_brand.iloc[:10]

fig, ax = plt.subplots()

ax.bar(
    x = df_quantity_brand_top10["Brand"],
    height = df_quantity_brand_top10["Quantity"],
    width = 0.6,
)

ax.set(
    title="Number of cars from top 10 most sold brands in Norway 2007-2017",
    xlabel="Car brand",
    ylabel="Number of cars",
)

fig.savefig("figures/bar_top_10_cars.png")
```

## Linechart

```python
df_quantity_year = duckdb.query("""
    SELECT Year, SUM(quantity) AS Quantity FROM df GROUP BY Year ORDER BY Year
""").df().set_index("Year")
```

```python
df_quantity_year.iloc[:-1].plot()
```

```python
ax = df_quantity_year.iloc[:-1].plot(title="Cars sales in Norway between 2007-2016", xlabel="Year", ylabel="Number Car Sold", style="--o", linewidth=2)

ax.legend().remove()
fig = ax.get_figure()
fig.tight_layout()
fig.savefig("figures/bar_quantity_year")
```
