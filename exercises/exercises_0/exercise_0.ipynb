### b) Draw a line chart of quantity for each year. Is there a year that should be skipped?

```python
import duckdb as db

df_year = db.query("""--sql
                   SELECT Year, SUM(Quantity) as Quantity,	
                   from df
                   GROUP BY Year
                   ORDER BY Year 
    """).df()

df_year
```

```python
df_year.iloc[:-1].plot(x="Year", y="Quantity")
```

 ### c) Draw a line chart of average CO2 emissions for same years that as in b)

```python
df.Avg_CO2
```

```python


df_co2_avg = db.query("""--sql
                   SELECT Year, SUM(Avg_CO2) as Avg_CO2,	
                   from df
                   GROUP BY Year
                   ORDER BY Year 
    """).df()

df_co2_avg
```

```python
df_co2_avg.iloc[:-1].plot(x="Year", y="Avg_CO2")
```

  ### d) Draw a line chart of all years and months for import

```python
df_month_make = db.query(
    """--sql
            SELECT Year || '-' || LPAD(CAST(Month AS VARCHAR),2,'0') AS date, 
                    SUM(Import) AS Import FROM df
            GROUP BY date
            ORDER BY date
    """
).df()

df_month_make
```

```python
df_month_make.iloc[:-1].plot(x="date", y="Import", figsize=(12,6))
```

###   e) Draw a line chart of all years and months for average CO2 emissions

```python
df_month_make_co2 = db.query(
    """--sql
            SELECT Year || '-' || LPAD(CAST(Month AS VARCHAR),2,'0') AS date, 
                    SUM(Avg_CO2) AS Avg_CO2 FROM df
            GROUP BY date
            ORDER BY date
    """
).df()

df_month_make_co2
```

```python
df_month_make_co2.plot(x="date", y="Avg_CO2", figsize=(12,6))
```

###   f) Draw a line chart of all years and months for electric cars import

```python
df_month_make_electric = db.query(
    """--sql
            SELECT Year || '-' || LPAD(CAST(Month AS VARCHAR),2,'0') AS date, 
                    SUM(Import_Electric) AS Import_Electric FROM df
            GROUP BY date
            ORDER BY date
    """
).df()

df_month_make_electric
```

```python
df_month_make_electric.plot(x="date", y="Import_Electric", figsize=(12,6))
```

### g) Draw a line chart of average diesel share per year

```python
df_diesel_avg = db.query("""--sql
                   SELECT Year, AVG(Diesel_Share) as Diesel,	
                   from df
                   GROUP BY Year
                   ORDER BY Year 
    """).df()

df_diesel_avg
```

```python
df_diesel_avg.iloc[:-1].plot(x="Year", y="Diesel", figsize=(12,6))
```
